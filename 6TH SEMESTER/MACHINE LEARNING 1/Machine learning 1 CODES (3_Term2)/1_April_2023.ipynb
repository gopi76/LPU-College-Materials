```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy = 'most_frequent')
df_si = si.fit_transform(df)
```

```python
df_si = pd.DataFrame(df_si)

df_si.isna().sum().sum()
```

```python
df_si.head()
```

```python
cols = [0,3,4,5,6,7,8,9,11,12,15]

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in cols:
    df_si[i] = le.fit_transform(df_si[i])
    
df_si.head()
```

```python
target = df_si[15]
data = df_si.drop(columns = [15] )
```

```python
target.shape
```

```python
data.shape
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data, target, test_size = 0.3)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
```

```python
pipe1 = Pipeline([('sc', StandardScaler()), ('pca', PCA()), ('svm1', SVC(kernel = 'rbf'))])
```

```python
pipe1.fit(x_train, y_train)
```

```python
train_pred = pipe1.predict(x_train)

test_pred  = pipe1.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Training accuracy : ", accuracy_score(train_pred, y_train))
```

```python
print("Testing accuracy : ", accuracy_score(test_pred, y_test))
```

```python
print("Testing accuracy : ", pipe1.score(x_test, y_test))
```

```python
print("Training accuracy : ", pipe1.score(x_train, y_train))
```

```python
from sklearn.model_selection import cross_validate
```

```python
for i in range(3,11):
    cv_results = cross_validate(pipe1, data, target, cv = i)
    print("K fold: ", i)
    print(np.average(cv_results['test_score']))
```

```python
print(np.average(cv_results['test_score']))
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
score = []
skf = StratifiedKFold(n_splits = 10)

for train_index, test_index in skf.split(data, target):
    x_train_fold, x_test_fold = data.iloc[train_index], data.iloc[test_index]
    y_train_fold, y_test_fold = target[train_index], target[test_index]
    
    pipe1.fit(x_train_fold, y_train_fold)
    score.append(pipe1.score(x_test_fold, y_test_fold))
```

```python
print(np.average(score))
```
